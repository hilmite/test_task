```python
import pandas as pd
```

```python
column_indices = [(35, 45), (53, 83), (106, 112), (137, 142), (155, 158)]

columns = ['indeks', 'nazwa', 'cena detaliczna netto', 'ilość w opakowaniu', 'grupa rabatowa']
df = pd.DataFrame(columns=columns)

file_path = 'PP0006_MULTI.txt'

with open(file_path, 'r') as file:
    for line in file:
        row_data = {}

        for i, (start, end) in enumerate(column_indices):
            column_name = f'Column{i + 1}'
            row_data[column_name] = str(line[start-1:end].strip())
        df.loc[len(df)] = row_data
        print(row_data)
df
```
